```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(100)

# Number of samples
n_samples = 100000

# Sample from the standard Cauchy distribution
samples = np.random.standard_cauchy(n_samples)

# Compute running mean
running_mean = np.cumsum(samples) / np.arange(1, n_samples + 1)

# Plot the evolution of the sample mean
plt.figure(figsize=(10, 5))
plt.plot(running_mean, label="Sample Mean")
plt.axhline(0, color='red', linestyle='--', label="Expected Mean (undefined)")
plt.title("Evolution of Sample Mean from a Cauchy Distribution")
plt.xlabel("Number of Samples")
plt.ylabel("Sample Mean")
plt.xscale('log')  # Logarithmic scale for x-axis
plt.legend()
plt.grid(True)
plt.ylim(-25, 25)  # Limit y-axis to better see divergence
plt.show()
```
